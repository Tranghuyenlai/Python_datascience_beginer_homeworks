```python
#3. Nhập 2 số nguyên a, b thực hiện các phép tính phần dư và phần nguyên của a cho b.
a = int(input("a: "))
b = int(input("b: "))
division = a / b
phan_nguyen = int(division)
phan_du = a % b
print("phần nguyên: " + str(phan_nguyen))
print("phần dư: " + str(phan_du))
```

```python
#4. Cho người dùng nhập vào năm sinh, in ra tuổi.
import datetime
date = datetime.date.today()
current_year = date.year
birth_year = int(input('Năm sinh: '))
age = current_year - birth_year
print ('I\'m ' + str(age) + " years old.")
```

```python
#5. Nhập vào chiều dài, chiều rộng của hình chữ nhật. Tính diện tích và chu vi
length = int(input("Chiều dài: "))
wide = int(input("Chiều rộng: "))
area = length * wide
perimeter = 2 * (length + wide)
print("Diện tích là " + str(area))
print("Chu vi là " + str(perimeter))
```

```python
#6. Nhập vào 2 số a và b. Tính a + b, a - b.
a = int(input("a: "))
b = int(input("b: "))
plus = a + b
minus = a - b
print(str(a) +' + ' + str(b) + ' = ' + str(plus))
print(str(a) +' - ' + str(b) + ' = ' + str(minus))
```

```python
#7. Cho người dùng nhập vào tổng số đôi tất trong ngăn kéo. In ra số lần rút trong trường hợp xấu nhất.
number = int(input("Tổng số đôi tất: "))
withdrawn = number + 1
print("Số lần rút trong trường hợp xấu nhất là " + str(withdrawn) + " lần.")
```

```python
#8. Input : Nhập vào tổng số người Output: In ra số lần bắt tay
number = int(input('Tổng số người: '))
shaking_hand = number*(number - 1)/2
print("Tổng số lần bắt tay là " + str(shaking_hand) + " lần.")
```
