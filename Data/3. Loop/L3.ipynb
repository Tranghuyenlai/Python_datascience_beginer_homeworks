cho nhập n, tính tổng từ 1->n

n = 10 -> 55

```python
total = 0
tich = 1

n = int(input('n = '))

for i in range(1, n+1):
  total += i
  tich *= i

# total = 1 + 2 + 3 + ... + n
print(total)
print(tich)
```

Số nguyên tố là số chỉ có đúng hai ước, là một và chính nó, 2, 3, 5, 7,...


```python
# 1, 2, .., 99 -> 99
for i in range(1, 100):
  if i % 3 == 0 and i % 5 != 0:
    print(i, end=' ')
```

```python
# 3, 6, 9, 12,..., 99 -> 33
for i in range(3, 100, 3):
  if i % 5 != 0:
    print(i, end=' ')
```
