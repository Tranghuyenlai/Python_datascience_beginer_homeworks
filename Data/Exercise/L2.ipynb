**6. Nhập a,b,c  giải và biện luận tìm nghiệm của phương trình ax^2 + bx + c = 0.**

```
#6. Nhập vào a,b,c. Giải và biện luận tìm nghiệm phương trình ax^2 + bx + c = 0.
import math 
a = float(input('a = '))
b = float(input('b = '))
c = float(input('c = '))
if a == 0:
  # bx + c = 0
  print('Với a = {0}, b = {1}, c = {2}, \nphương trình trở thành phương trình bậc nhất: bx + c = 0'.format(a,b,c)) 
  if b != 0:
    x = float((-c)/b)
    print('Ta có phương trình: {1}x + {2} = 0 <=> x = (-{2})/{1} <=> x = {3}'.format(a,b,c,x))
    print('Phương trình có nghiệm duy nhất x = {}'.format(x))
  elif b == 0:
    if c == 0:
      print('Ta có phương trình: \n{1}x + {2} = 0 <=> {1}x = {2}'.format(a,b,c))
      print('Kết luận: Phương trình có nghiệm đúng với mọi x')
    else:
      print('Ta có phương trình: \n{1}x + {2} = 0 <=> {1}x = -{2}'.format(a,b,c))
      print('Kết luận: Phương trình vô nghiệm')
if a != 0:
  delta = float(b**2 - 4*a*c)
  print('Với a = {0}, b = {1}, c = {2}, \nPhương trình trở thành {0}x^2 + {1}x + {2} = 0'.format(a,b,c))
  if delta < 0:
    print('Ta có delta = b^2 - 4ac => delta = ({1})^2 - 4*{0}*{2} = {3}'.format(a,b,c,delta))
    print('Do delta = {} <  0. Kết luận: Phương trình vô nghiệm'.format(delta))
  elif delta == 0:
    x = -b/(2*a)
    print('Ta có delta = b^2 - 4ac => delta = ({1})^2 - 4*{0}*{2} = {3}'.format(a,b,c,delta))
    print('Do delta = {0}. Kết luận: Phương trình lúc này có nghiệm kép x1 = x2 = {1}'.format(delta,x))
  else:
    sqrt_delta = math.sqrt(delta)
    x1 = ((-b) + float(sqrt_delta))/(2*a)
    x2 = ((-b) - float(sqrt_delta))/(2*a)
    print('Ta có delta = b^2 - 4ac = ({1})^2 - 4*{0}*{2} = {3}'.format(a,b,c,delta))
    print('Do delta = {} >  0. Kết luận: Phương trình có hai nghiệm phân biệt: '.format(delta))
    print('x1 = (-(b) + squareroot(delta))/2(a) = (-{1} + {2})/(2*{0}) = {3}'.format(a,b,sqrt_delta,x1))
    print('x2 = (-(b) - squareroot(delta))/2(a) = (-{1} - {2})/(2*{0}) = {3}'.format(a,b,sqrt_delta,x2))
```

**8. input độ dài 3 cạnh 1 tam giác, kiểm tra độ dài có hợp lệ không? nếu có, kiểm tra xem có phải tam giác đều/cân/vuông không?**

```
x = float(input('Length of side 1 = '))
y = float(input('Length of side 2 = '))
z = float(input('Length of side 3 = '))

# test case
# 1, 1, 3
# 3, 4, 5
# 3, 3, 4
# 5, 5, 5
if x + y > z and x + z > y and y + z > x:
  if x == y or x == z or y == z:
    if (x**2 == y**2 + z**2) or (z**2 == x**2 + y**2) or (y**2 == x**2 + z**2):
      print('It is an isosceles right triangle')
    else:
      print('It is an isosceles triangle')
  elif x == y and y == z:
    print('It is an equilateral triangle')  
  elif (x**2 == y**2 + z**2) or (z**2 == x**2 + y**2) or (y**2 == x**2 + z**2):
    print('It is a right triangle')
  else:
    print('It is an unknown triangle')
else:
  print('It is an unvalid triangle')
```

**9. input số điện output giá tiền, biết rằng: 50 số đầu tiên có giá 3.000/1 số, 100 số tiếp theo có giá 4.000/1 số, 100 số tiếp theo 5.000/1 số và từ 250 số trở đi thì giá là 10.000/1 số.**

```
so_dien = float(input('so dien: '))
# tien_1 = 3000
tien_1 = int(3000) #0<so_dien<=50
tien_2 = int(4000) #50<so_dien<=150
tien_3 = int(5000) #151 < so_dien <= 250
tien_4 = int(10000) # 250<= so_dien
if so_dien < 0:
  print('Số điện {} nhập vào không hợp lệ'.format(so_dien))
else:
  if 0 <= so_dien <= 50:
    tien_dien = so_dien*tien_1
  elif 50< so_dien <= 150:
    tien_dien = 50*tien_1 + (so_dien - 50)*tien_2
  elif 150< so_dien <= 250:
    tien_dien = 50*tien_1 + 100*tien_2 + (so_dien - 150)*tien_3
  elif so_dien > 250:
    tien_dien = 50*tien_1 + 100*tien_2 + 100*tien_3 + (so_dien - 250)*tien_4
  print('Với số điện {} đã sử dụng, tiền điện cần phải trả là {}'.format(so_dien, tien_dien))
```
