**5. Nhập vào chiều dài, chiều rộng của hình chữ nhật. Tính diện tích và chu vi**

```
# float: số thực
length= float(input("Chiều dài: "))
width = float(input("Chiều rộng: "))
dientich = length*width
chuvi= (length+width)*2
print(f"Diện tích hình chữ nhật: {dientich}")
print(f"Chu vi hình chữ nhật: {chuvi}")
```

**Cho người dùng nhập vào tổng số đôi tất trong ngăn kéo. In ra số lần rút trong trường hợp xấu nhất.** 

```
total_pairs_of_socks = int(input('Total pair of socks: '))
# the number of socks of different colors will be exactly the same as the number of pairs of socks
# so to have 1 pair of socks of the same color just pull 1 more sock
pull_out_a_sock_times = total_pairs_of_socks + 1
print('Number of times to pull out a sock in a drawer:', pull_out_a_sock_times)
```
